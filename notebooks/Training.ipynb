```python
import src
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
```

## Init Agent

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
model = nn.Sequential(
    nn.Linear(16, 16),
    nn.Hardsigmoid(),
    nn.Linear(16, 8),
    nn.Hardsigmoid(),
    nn.Linear(8, 4),
    nn.Hardsigmoid(),
    nn.Linear(4, 2),
    nn.Hardsigmoid(),
    nn.Linear(2, 1),
    nn.Hardsigmoid(),
)
loss_func = F.smooth_l1_loss
optimizer = optim.Adam(model.parameters())
value_func = src.value_func.NNFunc(model, loss_func, optimizer, device)
agent = src.Agent(value_func)
```

```python
%%time
for j in range(1000):
    play_batch = 1000
    history = src.History(play_batch, device)
    game = src.Game(play_batch, device=device, history=history)
    print(agent.play(game).float().mean())
    for i in range(100):
        sample = history.sample(100)
        agent.train(sample)
```
